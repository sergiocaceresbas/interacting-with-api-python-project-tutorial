```python
indicators = {                                             # Creamos un diccionario para relacionar cada código de la API con un nombre fácil de entender.
    "NY.GDP.PCAP.CD": "gdp_per_capita",                    
    "SP.DYN.LE00.IN": "life_expectancy",                   
    "EN.ATM.CO2E.PC": "co2_per_capita"                     
}
```

```python
def fetch_indicator_data(country_codes, indicator_code, start_year=2010, end_year=2024):
    # Creamos una función que descargará un indicador del Banco Mundial.

    country_path = ";".join(country_codes)                 # Unimos los códigos de países en un solo texto: ESP;FRA.

    url = f"https://api.worldbank.org/v2/country/{country_path}/indicator/{indicator_code}"
                                                            # Construimos la dirección concreta de la API.

    params = {                                              # Creamos las condiciones que acompañarán a la petición.
        "format": "json",                                   # Pedimos la respuesta en formato JSON.
        "date": f"{start_year}:{end_year}",                 # Indicamos el periodo de años.
        "per_page": 1000                                    # Pedimos hasta 1000 resultados en una página.
    }

    response = requests.get(url, params=params)             # Enviamos la petición a la API.

    response.raise_for_status()                             # Comprobamos que la petición no haya fallado.

    data = response.json()                                  # Convertimos la respuesta en datos de Python.

    return data                                             # Entregamos los datos descargados.
```

```python
datos_pib = fetch_indicator_data(                           # Ejecutamos la función 
    ["ESP", "FRA"],                                        #  España y Francia.
    "NY.GDP.PCAP.CD"                                       # Pedimos el PIB per cápita.
)

datos_pib[0]                                                # Mostramos los metadatos de la consulta.
```

```python
datos_pib[1][0]
```

```python
df = pd.DataFrame(datos_pib[1])
```

```python
df.head()
```

```python
df = df[
    [
        "country",
        "date",
        "value"
    ]
]
```

```python
df["country"] = df["country"].apply(
    lambda x: x["value"]
)
```

```python
df.columns = [
    "country",
    "year",
    "gdp_per_capita"
]
```

```python
df["year"] = df["year"].astype(int)
```

```python
df["gdp_per_capita"] = df["gdp_per_capita"].astype(float)
```

```python
df = df.sort_values(
    ["country", "year"]
)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))

for country in df["country"].unique():

    country_data = df[
        df["country"] == country
    ]

    plt.plot(
        country_data["year"],
        country_data["gdp_per_capita"],
        marker="o",
        label=country
    )

plt.title("PIB per cápita")
plt.xlabel("Año")
plt.ylabel("PIB per cápita")
plt.legend()
plt.grid(True)
plt.show()
```

## Análisis del gráfico del PIB per cápita

Podemos comprobar en el gráfico del PIB per cápita de ambos países que Francia presenta durante todo el periodo un PIB per cápita superior al de España, con una diferencia aproximada de unos 10.000 euros.

También observamos que ambos países siguen una evolución muy parecida: cuando el PIB per cápita aumenta o disminuye en uno de ellos, el otro suele mostrar una tendencia similar. Esto puede deberse a que ambos pertenecen a la zona euro y sus economías están muy relacionadas, aunque el gráfico por sí solo no permite confirmar la causa.

"Se observa una ligera disminución del PIB per cápita entre 2014 y 2016 en ambos países. El gráfico no permite determinar la causa, aunque podría estar relacionada con la evolución económica de la zona euro durante ese periodo."

Se aprecia un descenso importante alrededor del año 2020, que podría estar relacionado con la crisis provocada por la pandemia. A partir de ese momento, ambos países muestran una recuperación y en 2024 alcanzan algunos de los valores más altos del periodo analizado.
