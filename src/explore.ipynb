```python
for country in data[1]:
    print(country["name"], "-", country["capitalCity"])
```

```python
for country in data[1]:                                     # Recorre uno a uno todos los países de la lista.

    if country["capitalCity"] != "":                        # Si el país tiene una capital (no está vacía).

        print(country["name"], "-", country["capitalCity"]) # Muestra el nombre del país y su capital.

        
```

```python
paises = []                                                   # Creamos una lista vacía para guardar despues

for country in data[1]:                                       # Recorremos uno a uno todos los países que devuelve la API.

    if country["capitalCity"] != "":                          # Comprobamos que el país tenga una capital registrada.

        paises.append({                                       # Guardamos un nuevo diccionario solo con los datos que vamos a utilizar.
            "pais": country["name"],                          # Guardamos el nombre del país.
            "codigo": country["id"],                          # Guardamos el código del país.
            "capital": country["capitalCity"]                 # Guardamos la capital.
        })
len (paises)
```
